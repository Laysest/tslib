## Chattanooga data 

```python
import pandas as pd
from datetime import datetime
```

```python
df = pd.read_csv('./isolated-intersection/testbed-a-hour/2021-02-13.csv').sort_values(by='timestamp')
```

```python
# df = df[(df['timestamp'] > '20210213T160000') & (df['timestamp'] < '20210213T170000')].sort_values(by='timestamp')
df['timestamp'] = pd.to_datetime(df["timestamp"]) - pd.Timestamp('20210213T000000')
```

```python
def mapping_veh_class(length):
    l = length*0.3
    if l <= 1.8:
        return 'motorcycle'
    if l < 5.7:
        return 'passenger'
    if l < 7.2:
        return 'delivery'
    if l < 10:
        return 'truck'
    return 'semitrailer'
```

```python
def mapping_route(row):
    route = []
    if row['approach'] == 'W':
        route.append('WtoC')
        if row['turn'] == 'S':
            route.append('CtoE')
        elif row['turn'] == 'L':
            route.append('CtoN')
        elif row['turn'] == 'R':
            route.append('CtoS')
        else:
            route.append('CtoW')
    elif row['approach'] == 'E':
        route.append('EtoC')
        if row['turn'] == 'S':
            route.append('CtoW')
        elif row['turn'] == 'L':
            route.append('CtoS')
        elif row['turn'] == 'R':
            route.append('CtoN')
        else:
            route.append('CtoE')
    elif row['approach'] == 'N':
        route.append('NtoC')
        if row['turn'] == 'S':
            route.append('CtoS')
        elif row['turn'] == 'L':
            route.append('CtoW')
        elif row['turn'] == 'R':
            route.append('CtoE')
        else:
            route.append('CtoN')
    elif row['approach'] == 'S':
        route.append('StoC')
        if row['turn'] == 'S':
            route.append('CtoN')
        elif row['turn'] == 'L':
            route.append('CtoE')
        elif row['turn'] == 'R':
            route.append('CtoW')
        else:
            route.append('CtoS')
    return route[0] + ' ' + route[1] 
```

```python
for idx, row in df.iterrows():
    print("<vehicle id=\"{}\" depart=\"{}\" type=\"{}\" departSpeed=\"5\">\n \
   <route edges=\"{}\"/>\n \
</vehicle>\n".format(idx, row['timestamp'].seconds, mapping_veh_class(row['length_ft']), mapping_route(row)))
```
